### On Axis

```python
R = 2.398457
z = 0.39632
```

```python
HelmholtzCoil.FieldLoopOnAxis(z, R)
```

```python
print(HelmholtzCoil.FieldLoopBz(z, 0, R))
print(HelmholtzCoil.FieldLoopBrho(z, 0, R))
```

```python
HelmholtzCoil.FieldLoopDiscrete(z/R, 0)
```

```python
z = np.linspace(-3, 3, 31)
y1 = np.zeros_like(z)
y2 = np.zeros_like(z)
y3 = np.zeros_like(z)
for i, zi in enumerate(z):
    y1[i] = HelmholtzCoil.FieldLoopOnAxis(zi)
    y2[i] = HelmholtzCoil.FieldLoopBz(zi, 0)
    y3[i] = HelmholtzCoil.FieldLoopDiscrete(zi, 0, 5040)[2]
    
plt.plot(z, y1, 'x', label='FieldLoopOnAxis()')
plt.plot(z, y2, '+', label='FieldLoopBz()' )
plt.plot(z, y3, '1', label='FieldLoopDiscrete()')
plt.legend()
plt.grid()
plt.show()

plt.plot(z, (y2-y1)/y1, '+')
plt.plot(z, (y3-y1)/y1, '1')
plt.grid()
plt.show()
```
